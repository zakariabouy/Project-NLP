```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Exemple simple
corpus = [
    "J'aime la programmation Python ",
    "Python est un langage puissant",
    "J'aime apprendre de nouvelles choses"
]
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
```

```python
print("Matrice de caractéristiques:")
print(X.toarray())
print("\nVocabulaire:")
print(vectorizer.get_feature_names_out())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Pipeline de classification
corpus = ["texte positif", "texte négatif", "excellent produit"]
labels = [1, 0, 1]  # 1=positif, 0=négatif

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)

# Entraînement d'un modèle
classifier = MultinomialNB()
classifier.fit(X, labels)

# Prédiction sur nouveau texte
new_text = ["produit excellent"]
X_new = vectorizer.transform(new_text)
predictions = classifier.predict(X_new)
```
